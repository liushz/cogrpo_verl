```python
import numpy as np

def mc_variance_bias(trials=20000, n=5, true_sigma=2.0, seed=0):
    rng = np.random.default_rng(seed)
    # 从 N(0, sigma^2) 抽样
    X = rng.normal(loc=0.0, scale=true_sigma, size=(trials, n))
    xbar = X.mean(axis=1, keepdims=True)
    s2_n = ((X - xbar)**2).mean(axis=1)              # 除以 n
    s2_n1 = ((X - xbar)**2).sum(axis=1) / (n - 1)    # 除以 n-1
    
    true_var = true_sigma**2
    return {
        "true_var": true_var,
        "E[s2_n]": s2_n.mean(),
        "bias(s2_n)": s2_n.mean() - true_var,
        "E[s2_n1]": s2_n1.mean(),
        "bias(s2_n1)": s2_n1.mean() - true_var,
        "Var(s2_n)": s2_n.var(),
        "Var(s2_n1)": s2_n1.var(),
    }

print(mc_variance_bias(n=5))
print(mc_variance_bias(n=50))


```
